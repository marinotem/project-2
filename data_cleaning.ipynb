```python
#adding in dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# adding homicide file as a csv in pandas
csv_file = "Resources/homicide.csv"
homicide_df = pd.read_csv(csv_file)
homicide_df.head()
```

```python
# cleaning homicide df 
# taking off agency id, agency name, agency type 
homicide_df = homicide_df.drop(columns=["Agency Code", "Agency Name", "Agency Type", "State"])
homicide_df
```

```python
# cleaning out only for cities that are in our weather datasets
homicide_clean_df = homicide_df[homicide_df['City'].isin(['Portland','San Francisco','Seattle','Los Angeles','San Diego','Las Vegas','Phoenix','Albuquerque','Denver','San Antonio','Dallas','Houston','Kansas City','Minneapolis','Saint Louis','Chicago','Nashville','Indianapolis','Atlanta','Detroit','Jacksonville','Charlotte','Miami','Pittsburgh','Toronto','Philadelphia','New York','Boston'])]
homicide_clean_df 
```

```python
#using group by to aggregate homicides by month for each city
homicide_cleaner_df = homicide_clean_df.groupby(['City','Year','Month'],as_index=False)['Record ID'].agg('count')
homicide_cleaner_df
```

```python
# import temp csv as df 
csv_file = "Resources/temperature.csv"
temp_df = pd.read_csv(csv_file)
temp_df.head()
```

```python
# import humdiity csv as df 
csv_file = "Resources/humidity.csv"
humidity_df = pd.read_csv(csv_file)
humidity_df.head()
```

```python
# import weather event csv as df 
csv_file = "Resources/weather_description.csv"
weather_event_df = pd.read_csv(csv_file)
weather_event_df.head()
```

```python
#drop nans from each df 
temp_clean_df = temp_df.dropna()
humidity_clean_df = humidity_df.dropna()
weather_event_clean_df = weather_event_df.dropna()

temp_clean_df
```

```python
#making the column names become rows in a column, this melt function can only be ran once or else it keeps going
temp_cleanest_df = temp_clean_df.melt(id_vars="datetime")
temp_cleanest_df
```

```python
#making the column names become rows in a column, this melt function can only be ran once or else it keeps going
humidity_clean_df = humidity_clean_df.melt(id_vars="datetime", 
        var_name="City", 
        value_name="Humidity")
```

```python
#making the column names become rows in a column, this melt function can only be ran once or else it keeps going
weather_event_clean_df = weather_event_clean_df.melt(id_vars="datetime", 
        var_name="City", 
        value_name="Weather Event")  
weather_event_clean_df   
```

```python
# separating out month and year from datetime
temp_cleanest_df['month'] = temp_cleanest_df["datetime"].str[5:7]
temp_cleanest_df['year'] =temp_cleanest_df["datetime"].str[0:4]

weather_event_clean_df['month'] = weather_event_clean_df["datetime"].str[5:7]
weather_event_clean_df['year'] = weather_event_clean_df["datetime"].str[0:4]

humidity_clean_df['month'] = humidity_clean_df["datetime"].str[5:7]
humidity_clean_df['year'] = humidity_clean_df["datetime"].str[0:4]
```

```python
#calling in calendar dependency to help do datetime conversion
import calendar
```

```python
#converting numeric month to month name as a string
temp_cleanest_df['month'] = pd.to_datetime(temp_cleanest_df['month'], format='%m').dt.month_name()

humidity_clean_df['month'] = pd.to_datetime(humidity_clean_df['month'], format='%m').dt.month_name()

weather_event_clean_df['month'] = pd.to_datetime(weather_event_clean_df['month'], format='%m').dt.month_name()

```

```python
#converting year column from object variable to an integer
temp_cleanest_df["year"] = pd.to_numeric(temp_cleanest_df["year"])
humidity_clean_df["year"] = pd.to_numeric(humidity_clean_df["year"])
weather_event_clean_df["year"] = pd.to_numeric(weather_event_clean_df["year"])
```

```python
#filtering data table to only include years previous or equal to 2014
temp_cleanest_df =temp_cleanest_df.loc[temp_cleanest_df["year"] <= 2014 ]
humidity_clean_df = humidity_clean_df.loc[humidity_clean_df["year"] <= 2014 ]
weather_event_clean_df = weather_event_clean_df .loc[weather_event_clean_df["year"] <= 2014 ]
humidity_clean_df
```

```python
#renaming some columns for ease
temp_cleanest_df.rename(columns = {'variable':'City', 'value':'Temperature', 'month': 'Month', 'year':'Year'}, inplace = True)
humidity_clean_df.rename(columns = {'month': 'Month', 'year':'Year'}, inplace = True)
weather_event_clean_df.rename(columns = {'month': 'Month', 'year':'Year', 'Weather Event': "Most Common Weather Event"}, inplace = True)
```

```python
#Getting the median of temperature and humidity for each month for each city in each year
temp_cleanest_df = temp_cleanest_df.groupby(['City','Month', 'Year'],as_index=False)['Temperature'].median()
humidity_clean_df = humidity_clean_df.groupby(['City','Month', 'Year'],as_index=False)['Humidity'].median()
```

```python
# getting the mode for most common weather events as it is a string
#split on column and take first value 
weather_event_cleanest_df = weather_event_clean_df.groupby(['City','Month', 'Year'],as_index=False)['Most Common Weather Event'].agg(pd.Series.mode)
weather_event_cleanest_df
```

```python
#changing to string type and checking values
weather_event_cleanest_df['Most Common Weather Event'] = weather_event_cleanest_df['Most Common Weather Event'].astype(str)
weather_event_cleanest_df['Most Common Weather Event'].value_counts()
```

```python
# removing square brackets from items in list in column where there were multiple modes
weather_event_cleanest_df.loc[:, 'test'] = weather_event_cleanest_df['Most Common Weather Event'].apply(lambda x: x.split("' '")[0].replace("['",""))
```

```python
weather_event_cleanest_df['test'].value_counts()
```

```python
#checking data
weather_event_cleanest_df.head()
```

```python
#dropping repeat columns
weather_event_cleanest_df.drop(columns='Most Common Weather Event')
```

```python
# merging together the weather event df, temp df, humidity df and homicide dfs 
homicide_weather_df = pd.merge(homicide_cleaner_df, temp_cleanest_df, on=['City','Month', 'Year']).merge(humidity_clean_df, on=['City','Month', 'Year']).merge(weather_event_cleanest_df, on=['City','Month', 'Year'])
homicide_weather_df
```

```python
# convert Temperature column from Kelvin to Fahrenheit
homicide_weather_df["Temperature (F)"] = (homicide_weather_df["Temperature"]  - 273.15) * (9 / 5) + 32
homicide_weather_df
```

```python
#drop extra temperature column now that we converted kelvin to fahrenheit
homicide_weather_df = homicide_weather_df.drop(columns=["Temperature"])
homicide_weather_df = homicide_weather_df.drop(columns=["Most Common Weather Event"])
homicide_weather_df
```

```python
# changing temperature (F) to just temperature because we ran into an issue with importing df to sql with a parentheses
homicide_weather_df.rename(columns = {'Temperature (F)':'Temperature'}, inplace = True)
homicide_weather_df
```

```python
#renaming columns to match sql
homicide_weather_df.rename(columns = {'City':'city', 'Year': 'year', 'Month': 'month', 'Record ID':'record_id','Humidity':'humidity','test':'most_common_weather_event', 'Temperature': 'temperature'}, inplace = True)
```
